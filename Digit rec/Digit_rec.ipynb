### Model Building

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import BinaryCrossentropy, SparseCategoricalCrossentropy
```

```python
X_train = df_train.drop(["label"],axis=1)
X_train.head()
```

```python
Y_train = df_train[["label"]]
Y_train.head()
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_cv, Y_tr, Y_cv = train_test_split( X_train, Y_train, test_size=0.3, random_state=4)
```

```python
X_tr = X_tr.to_numpy()
X_cv = X_cv.to_numpy()
Y_tr = Y_tr.to_numpy()
Y_cv = Y_cv.to_numpy()
```

```python
l1 = Dense(units = 25, activation = "relu")
l2 = Dense(units = 15, activation = "relu")
l3 = Dense(units = 10, activation = "softmax")
```

```python
model = Sequential([l1,l2,l3])
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss = SparseCategoricalCrossentropy())
model.fit(X_tr,Y_tr,epochs=100)
```

```python
cv_predict = model.predict(X_cv)
```

```python
cv_pr = []
for i in cv_predict:
    cv_pr.append(np.argmax(i))
```

```python
cv_pr
```

```python
X_te = df_test.to_numpy()
```

```python
test_predict = model.predict(X_te)
```

```python
te_pr = []
for i in test_predict:
    te_pr.append(np.argmax(i))
```

```python
test_predictions = pd.DataFrame(te_pr)
test_predictions.columns = ["Label"]
test_predictions.head()
```

```python
ids = [i+1 for i in range(len(te_pr))]
```

```python
ids = pd.DataFrame(ids)
```

```python
ids.columns = ["ImageId"]
```

```python
ids.head()
```

```python
final = pd.concat([ids,test_predictions],axis=1)
final.head()
```

```python
final.to_csv("final1.csv",index=False)
```
